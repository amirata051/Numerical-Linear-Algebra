```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
A = np.array([[1.0, 1/2, 1/3],
              [1/2, 1/3, 1/4],
              [1/3, 1/4, 1/5]])
A_inv = np.linalg.inv(A)
print("A_inv :\n",A_inv)

X = A_inv @ np.ones(3)
print("X :\n",X)
```

```python
A_deltaA = np.array([[1.0, 1/2, 1/3+0.00003],
              [1/2, 1/3, 1/4],
              [1/3, 1/4, 1/5]])
A_deltaA_inv = np.linalg.inv(A_deltaA)
print("(A + delta(A))_inv :\n",A_deltaA_inv)

X_hat = A_deltaA_inv @ np.ones(3)
print("X_hat :\n",X_hat)
```

```python
deltaX = X_hat - X
print("delta X :\n",deltaX)

print(f"norm1(delta X) / norm1(X) = {np.linalg.norm(deltaX, ord=1)} / {np.linalg.norm(X, ord=1)} = {np.linalg.norm(deltaX, ord=1)/np.linalg.norm(X, ord=1)}")
```

```python
A = np.array([[-0.001, -0.01, -0.01],
              [2*0.001, 0 ,-0.0001],
              [1, 5*0.001, 4]])

D = np.diag([1/np.linalg.norm(A[i,:], ord=2) for i in range(len(A))])
print("D = \n",D)

DA = D @ A
print("DA = \n", DA)
```

```python
ATA = A.T @ A
print("A_t.A = \n",ATA)
eigenvalues, eigenvectors = np.linalg.eig(ATA)
print(f"Condition number(A_t.A)2 = sqrt({max(eigenvalues)} / {min(eigenvalues)}) = {math.sqrt(max(eigenvalues)/min(eigenvalues))}")
```

```python
DATDA = DA.T @ DA
print("DA_t.DA = \n",DATDA)
eigenvalues = np.abs(np.linalg.eigvals(DATDA))
print(f"Condition number(DA_t.DA)2 = sqrt({max(eigenvalues)} / {min(eigenvalues)}) = {math.sqrt(max(eigenvalues)/min(eigenvalues))}")
```

```python
def hilbert(n):
    hilbert = np.zeros((n,n))
    for i in range(1,n+1):
      for j in range(1,n+1):
        hilbert[i-1,j-1] = 1/(i+j-1)

    return hilbert
```

```python
def condition_number(A):
    eigenvalues = np.abs(np.linalg.eigvals(A.T@A))
    condition_num = math.sqrt(max(eigenvalues) / (min(eigenvalues)))
    return condition_num
```

```python
data_x, data_y = [], []

for size in range(1, 10):
    data_x.append(size)
    condition_value = condition_number(hilbert(size))
    data_y.append(condition_value)
    plt.text(size, condition_value, f'{condition_value:.2f}', fontsize=9, ha='center', va='bottom')

plt.plot(data_x, data_y, marker='o', linestyle='--')
plt.xlabel('Size(n)')
plt.ylabel('Condition Number')
plt.title('Variation of condition number across different sizes of Hilbert matrix')
plt.grid(True)
plt.show()
```
